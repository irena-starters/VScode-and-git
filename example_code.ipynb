## Visualizations using seaborn

In the following example, we will plot some data using the seaborn library. This library doesn't come pre-installed with python, so one should install it in the terminal. Simply go to **Terminal** -> **New terminal** and type: 
> pip install seaborn   
 
Then, you can import the package with the following code:

```python
import seaborn as sns
%matplotlib inline   
sns.set_theme(style="whitegrid")
```

Since seaborn comes with buit-in datasets, we can load that that and start plotting it. We will load a data frame called tips.

```python
tips = sns.load_dataset('tips')
tips.head()
```

The seaborn package is a data visualization package and it contains many pre-buit plotting options. For example using boxplot() function, one can visualize the data with boxplots and using violinplot() function, one can visualize the data with violin plots. For example:

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='pastel')
```

Differnt examples of what can be plotted using seaborn package, can be found [here](https://seaborn.pydata.org/examples/index.html).
